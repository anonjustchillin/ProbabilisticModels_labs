```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(0, 8, 100)

f1= t-pow(t, 2)
f2= 1+t
f3 = -pow(t, 2)-t+4
f4 = 5-t

plt.plot(t, f1, color='r', label='X_1(t)')
plt.plot(t, f2, color='b', label='X_2(t)')
plt.plot(t, f3, color='g', label='X_3(t)')
plt.plot(t, f4, color='c', label='X_4(t)')

plt.legend()
plt.show()
```

```python
plt.plot(t, f2, color='b', label='X_max')
plt.plot(t, f3, color='g', label='X_min')
plt.legend()
plt.show()
```

```python
mx_t = 3.5-0.5*t-0.5*pow(t, 2)
plt.plot(t, f2, color='b', label='X_max')
plt.plot(t, mx_t, color='purple', label='M_x(t)')
plt.plot(t, f3, color='g', label='X_min')
plt.legend()
plt.show()
```
